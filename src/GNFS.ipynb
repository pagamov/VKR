## f1

```python
class F1:
    def __init__(self,n,m,k):
        self.m = m
        self.k = k
        self.res = []
        while k >= 0:
            self.res.append(n // m**k)
            n = n - (n // m**k) * m**k
            k = k - 1
    def __call__(self,x):
        y = 0
        for i in range(len(self.res)):
            y += self.res[i]*(x**(self.k-i))
        return y
f1 = F1(n,m,3)
f1.res
```

## Рациональная факторная база

```python
B1 = 30
FB2 = range(2,B1+1)
```

```python
def prime(num):
    if num < 2:
        return False
    for i in range(2,num):
        if num % i == 0:
            return False
    return True
def make_primes(arr):
    res = []
    for i in range(len(arr)):
        if prime(arr[i]):
            res.append(arr[i])
    return res
```

```python
FB2 = make_primes(FB2)
FB2
```

```python
def NOD(a,b):
    while a != 0 and b != 0:
        if a > b:
            a = a % b
        else:
            b = b % a
    return a+b
```

```python
def Nr(a,b):
    return round(b**(f1.k) * f1(a/b))

def SplitSp(n, FB):
    for p in FB:
        while n % p == 0 and n != 0:
            n //= p
    if abs(n) == 1:
        return True
    return False

def SplitSpAl(n, FB):
    for p in FB:
        while n % p[0] == 0 and n // p[0] != 0:
            n //= p[0]
    if abs(n) == 1:
        return True
    return False

def SP(L1,L2,n):
    res = []
    b = 1
    while True:
        for a in range(-L1,L1+1):
            flag = False
            for p in FB1[15:]:
                if (a - b * p[1]) % p[0] == 0:
                    flag = True
                    break
            if flag:
                if SplitSp(a - m * b, FB2) and SplitSpAl(Nr(a,b),FB1) and math.gcd(a,b) == 1:
                    print(f"({a},{b})", end=' ')
                    res.append([a,b])
            if len(res) == n:
                return res
        b += 1
    return res
```

## Вторичное просеивание

```python
def SecSP(arr, FB2):
    top_row = ['-1'] + [str(i) for i in FB2]
    res = []
    side = []
    for p in arr:
        r = [0] * (len(FB2)+1)
        y = p[0] - m * p[1]
        if y < 0:
            r[0] = 1
            y = y * -1
        for i in range(len(FB2)):
            while y % FB2[i] == 0 and y // FB2[i] != 0:
                y //= FB2[i]
                r[i+1] += 1
        
        res.append(r)
        side.append('('+str(p[0])+', '+str(p[1]) + '): ' +  str(p[0] - m*p[1]))
    return [top_row,side,res]

def Display(struct):
    import pandas as pd
    import numpy as np
    a = np.array(struct[2])
    df = pd.DataFrame(a, columns=struct[0], index=struct[1])
    print(df, df.shape)
```

```python
def SecSPA(arr, FB1):
    top_row = [str(i[0]) for i in FB1]
    res = []
    side = []
    for p in arr:
        r = [0] * (len(FB1))
        y = Nr(p[0],p[1])
        if y < 0:
            y = y * -1
        for i in range(len(FB1)):
            while y % FB1[i][0] == 0 and y // FB1[i][0] != 0:
                y //= FB1[i][0]
                r[i] += 1
        
        res.append(r)
        side.append('('+str(p[0])+', '+str(p[1]) + '): ' +  str(Nr(p[0],p[1])))
    return [top_row,side,res]
```

## Значения квадратичного характера

```python
def legendre(a, p):
    if pow(a, (p - 1) // 2) % p == 1:
        return 1
    else:
        return 0
```

```python
def QuadraticHar(smooth, FB3):
    res = []
    side = []
    top_row = ['('+str(i[0])+', '+str(i[1])+')' for i in FB3]
    for sm in smooth:
        side.append('('+str(sm[0])+', '+str(sm[1]) + ')')
        r = []
        for qh in FB3:
            a = legendre((sm[0] - sm[1] * qh[1]), qh[0])
            r.append(a)
        res.append(r)
    return [top_row,side,res]
```

```python
QH = QuadraticHar(smooth, FB3)
```

```python
Display(QH)
```

```python
def makeMatrix(smooth,s,ss,QH):
    matrix = []
    b = []
    for i in range(len(smooth)):
        matrix.append(s[2][i] + ss[2][i] + QH[2][i])
        b.append([smooth[i]])
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            matrix[i][j] %= 2
    return matrix, b
```

```python
def solve(m,b):
    for j in range(len(m[0])):
        for i in range(len(m)):
            if m[i][j] == 1:
                # нашли в j столбце i элемент
                for row in range(len(m)):
                    if row != i and m[row][j]:
                        for col in range(j,len(m[0])):
                            m[row][col] = (m[row][col] + m[i][col]) % 2
                        b[row].append(b[i][0])
    return m,b
```

```python
mn,bn = solve(m,b)
```
